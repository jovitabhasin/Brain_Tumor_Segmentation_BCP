### Some assignment problems

```python
# Assignment 1
'''
Import videogamesales.csv and perform the following actions on it
Add a column of 'global_sales' showing the total sales of all the different regions to the data frame and sort (highest first) and print the DataFrame according to it.
Display a plot of the total number of copies sold of each genre globally.
Filter out only the games containing 'Grand Theft Auto' in their name and display the following as a DataFrame:
a. their name
b. the platform they were released on
c. the year they were released in
d. the sum of sales in only Europe and Japan
Display a pie chart of the total sales of all Grand Theft Auto games combined in North America, Europe, Japan, and other sales.
'''
```

```python
df = pd.read_csv("videogamesales.csv")
df["Global_Sales"] = df["NA_Sales"] + df["EU_Sales"] + df["JP_Sales"] + df["Other_Sales"]
df = df.sort_values(by="Global_Sales", ascending=False)
display(df.head(10))
df.shape
```

```python
# Display a plot of the total number of copies sold of each genre globally.
genre_sales = df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)
print(genre_sales.head())
plt.figure(figsize=(12, 7))
genre_sales.plot(kind='bar', color='skyblue')
plt.title('Total Global Sales by Genre')
plt.xlabel('Genre')
plt.ylabel('Total Global Sales (millions)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Filter out only the games containing 'Grand Theft Auto' in their name
gta_games = df[df['Name'].str.contains('Grand Theft Auto', na=False)]

# Calculate the sum of sales in Europe and Japan for these games
gta_games_sales = gta_games.copy()
gta_games_sales['EU_JP_Sales'] = gta_games_sales['EU_Sales'] + gta_games_sales['JP_Sales']

# Display the required columns as a DataFrame
display(gta_games_sales[['Name', 'Platform', 'Year', 'EU_JP_Sales']])
```

```python
# Calculate total sales for each region for Grand Theft Auto games
na_sales_gta = gta_games_sales['NA_Sales'].sum()
eu_sales_gta = gta_games_sales['EU_Sales'].sum()
jp_sales_gta = gta_games_sales['JP_Sales'].sum()
other_sales_gta = gta_games_sales['Other_Sales'].sum()

# Prepare data for the pie chart
sales_data = [na_sales_gta, eu_sales_gta, jp_sales_gta, other_sales_gta]
sales_labels = ['North America', 'Europe', 'Japan', 'Other Regions']

# Create the pie chart
plt.figure(figsize=(10, 8))
plt.pie(sales_data, labels=sales_labels, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Total Sales of Grand Theft Auto Games by Region')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
